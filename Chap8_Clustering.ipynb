#K-means

### Generate Data

```python
from sklearn.datasets import make_blobs  #blobs คือ data ที่เป็นกลุ่ม 
```

```python
X, y = make_blobs(n_samples=500, centers=3, n_features=2,random_state=6)  #มี n = 500 ตัว    center มี 3 กลุ่ม  features =2 จะได้สร้างกราฟง่ายๆ
```

###Explore data

```python
X.shape
```

```python
from matplotlib import pyplot as plt
plt.scatter(X[:,0],X[:,1])
```

##Clustering

###import

```python
from sklearn.cluster import KMeans
```

### Define

```python
kmeans = KMeans(n_clusters=2, random_state=6)  #cluster มี 2 กลุ่ม 
```

###Train-Predict

```python
y_clust = kmeans.fit_predict(X)   #X  คำนวณหากลุ่ม
```

```python
y_clust
```

```python
plt.scatter(X[:,0],X[:,1],c=y_clust)   #y คือสีเหลือง
```

### 3 clusters

```python
#define
kmeans = KMeans(n_clusters=3, random_state=6)  #cluster มี 3 กลุ่ม 
y_clust = kmeans.fit_predict(X)   #X  คำนวณหากลุ่ม
#plot
plt.scatter(X[:,0],X[:,1],c=y_clust)   #y คือสีเหลือง
```

### 4 clusters

```python
#define
kmeans = KMeans(n_clusters=4, random_state=6)  #cluster มี 4 กลุ่ม 
y_clust = kmeans.fit_predict(X)   #X  คำนวณหากลุ่ม
#plot
plt.scatter(X[:,0],X[:,1],c=y_clust)   #y คือสีเหลือง
```

###5 clusters

```python
#define
kmeans = KMeans(n_clusters=5, random_state=6, init = 'random')  #cluster มี 4 กลุ่ม 
y_clust = kmeans.fit_predict(X)   #X  คำนวณหากลุ่ม
#plot
plt.scatter(X[:,0],X[:,1],c=y_clust)   #y คือสีเหลือง
```

```python
#Define
kmeans = KMeans(n_clusters=5, random_state=6)
#Fit-Predict
y_clust = kmeans.fit_predict(X)
#plot
plt.scatter(X[:,0],X[:,1], c=y_clust)
```

###Example  Application (Color Quantization)

```python
import cv2
import os
```

```python
img = cv2.imread('/content/25-images-of-smiling-faces-of-animals-that-make-you-smile2.jpg')
```

```python
img.shape
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

###นับจำนวนสี

```python
import numpy as np
```

```python
np.unique(np.reshape(RGB_img,(RGB_img.shape[0]*RGB_img.shape[1],3)),axis=0).shape #มี 126658 สี
```

####จัดกลุ่มสีให้เหลือ 16 สี

```python
myquantize = KMeans(n_clusters=16, random_state=6)
```

แปลงข้อมูลให้อยู่ในรูป row-column

```python
img_row = np.reshape(RGB_img,(RGB_img.shape[0]*RGB_img.shape[1],3))
img_row.shape
```

```python
final_color = myquantize.fit_predict(img_row)
```

ใช้ centroid เป็นตัวแทนของสี

```python
centroid = myquantize.cluster_centers_
```

```python
centroid
```

แทนสีคืนลงไป

```python
new_img = np.zeros(RGB_img.shape)
for i in range(RGB_img.shape[0]):
    for j in range(RGB_img.shape[1]):
      new_img[i,j,:] = centroid[myquantize.predict([RGB_img[i,j,:]]),:]
```

```python
new_img
```

```python
plt.imsave('img.jpg',new_img/255)
```
